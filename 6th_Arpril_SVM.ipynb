Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in
SVM?

```python
## Hyperplane : A hyperplane is a decision boundary that separates the data into two classes. In SVM, a hyperplane is a linear function that separates 
## the positive and negative examples with the maximum margin. The hyperplane is defined as w^T x + b = 0, where w is the weight vector, b is the bias term,
## and x is the input vector.
```

```python
## Marginal Plane: The marginal plane is the plane that is equidistant from the positive and negative examples and parallel to the hyperplane.
```

```python
## Hard Margin: A hard margin SVM is a type of SVM where we enforce that all training examples are correctly classified and lie outside the margin.
## In other words, we require that there are no misclassifications, and the margin is maximized.
```

```python
## Soft Margin: A soft margin SVM is a type of SVM where we allow some misclassifications to occur to improve the generalization performance of the classifier.
## We introduce a penalty term that trades off between the margin size and the misclassification error.
```

Q6. SVM Implementation through Iris dataset.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
svc=SVC()
```

```python
#pip install mlxtend
```

```python
from sklearn import datasets
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
data=datasets.load_iris()
```

```python
data.feature_names
```

```python
data.target
```

```python
data.target_names
```

```python
X=data.data[:,[0,2]]
```

```python
X[:5]
```

```python
y=data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
svm_model = SVC(kernel='linear', C=1, decision_function_shape='ovr')
```

```python
svm_model.fit(X_train, y_train)
```

```python
y_pred=svm_model.predict(X_test)
print("Accuracy Score  {} ".format(accuracy_score(y_test,y_pred)))
print("Confusion matric :\n",confusion_matrix(y_test,y_pred))
print("Classification Report :\n",classification_report(y_test,y_pred))
```

```python
plot_decision_regions(X_test, y_test, clf=svm_model, legend=2)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter={'kernel':['linear'],'C':[.2,.5,1,5,10,20]}
```

```python
grid_svc=GridSearchCV(estimator=svc,param_grid=parameter,cv=5,verbose=3)
```

```python
grid_svc.fit(X_train, y_train)
```

```python
grid_svc.best_params_
```

```python
grid_svc.best_score_
```
